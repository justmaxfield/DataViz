```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#Dataset for braodway gross revenue and number of weeks for every show over the past 8 years, broken up by type: Musical, Play, or Special Performance
df = pd.read_csv('Broadway.csv')

df.head()
```

```python
#Set style
plt.style.use('seaborn-whitegrid')

#Create empty canvas
fig = plt.figure()

#Add axes to canvas
axes1 = fig.add_axes([0.1,.1,.8,.8])
axes2 = fig.add_axes([.19,.5,.3,.3])

#Main plot
axes1.set_xlabel('Number of Weeks Running')
axes1.set_ylabel('Gross Revenue (in hundreds of millions)')
axes1.set_xlim(0,450)
axes1.set_ylim(0,800000000)
axes1.scatter(df['Mweek'],df['Mgross'], s = df['Mgross'] / 1000000, alpha = .65, color = 'b')
axes1.scatter(df['Pweek'], df['Pgross'], s = df['Pgross'] / 1000000, alpha = .65, color = 'r')
axes1.scatter(df['Sweek'], df['Sgross'], s = df['Sgross'] / 1000000, alpha = .65, color = 'g')

#Inset plot
axes2.scatter(df['Mweek'],df['Mgross'], s = df['Mgross'] / 1000000, alpha = .65, color = 'b')
axes2.scatter(df['Pweek'], df['Pgross'], s = df['Pgross'] / 1000000, alpha = .65, color = 'r')
axes2.scatter(df['Sweek'], df['Sgross'], s = df['Sgross'] / 1000000, alpha = .65, color = 'g')
axes2.set_xlim(0,52)
axes2.set_ylim(0,60000000)
axes2.set_ylabel('(in tens of millions)')
axes2.set_title('One Year')

#Format legend
axes1.legend(('Musicals', 'Plays', 'Specials'), loc="upper left", bbox_to_anchor=(1,1))

```

```python
#Interpretation:  Musicals make up the majority of revenue on broadway.  
#Multiple shows extend past the maximum 399 weeks that this dataset covers.  
#Only two plays have run longer than one year
#While Musicals can be high grossing and long running, there are still many that end in under 1 year.
#Special performances do not outperform plays or musicals of similar run lengths
```
